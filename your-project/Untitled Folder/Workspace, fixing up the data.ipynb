```python
# We want to merge income per country taxes and happiness 
import pandas as pd
import numpy as np
```

```python
World_happiness = pd.read_csv("Data_project/World_happiness/2018.csv")
GDP_per_capita = pd.read_excel("Data_project/GDP_per_capita.xlsx")
Taxes = pd.read_csv("Data_project/taxes4.csv")
```

```python
World_happiness.rename({"Country or region" : "Country"}, axis="columns",inplace=True)
#World_happiness
```

```python

GDP_per_capita.rename({2018 : "GDP_per_capita"}, axis="columns",inplace=True)
GDP_per_capita
```
